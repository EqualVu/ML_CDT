Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the data
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.rand(100, 1)

# Create the scatter plot
plt.scatter(x, y, color='blue', label='Data points',s=10)

# Add title and labels
plt.title('Scatter Plot of x and y')
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the data
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.rand(100, 1)

# Sort the data by x
sorted_indices = np.argsort(x, axis=0).flatten()
x_sorted = x[sorted_indices]
y_sorted = y[sorted_indices]

# Create the line plot
plt.plot(x_sorted, y_sorted, color='blue', label='Data line')

# Add title and labels
plt.title('Line Plot of x and y')
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

Tính theta bằng phương trình pháp tuyến sao cho mất mát đạt cực tiểu



```python
x_b = np.c_[np.ones((100, 1)),x] # add x0 = 1 to each instance
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)

print('array (',theta_best,')')
```

Thực hiện dự đoán với theta

```python
x_new = np.array([[0],[2]])
x_new_b = np.c_[np.ones((2,1)),x_new] # add x0=1 to each instance
y_predict = x_new_b.dot(theta_best)
y_predict
```

Vẽ đồ thị mô hình dự đoán

```python
plt.plot(x_new, y_predict, "r-",label='True Line')
plt.plot(x,y,"b.",label='Data Points')
plt.axis([0,2,0,15])
plt.legend()
plt.show()
```
